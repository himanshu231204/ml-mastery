## 1. Import & Data Structures

```python
import pandas as pd
import numpy as np

print("Pandas version:", pd.__version__)
```

Pandas has two core data structures:
- **Series** — a 1D labeled array (like a single column)
- **DataFrame** — a 2D labeled table (like a spreadsheet or SQL table)


```python
# Series
s = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
print("Series:\n", s)
print("\nIndex:", s.index.tolist())
print("Values:", s.values)          # underlying NumPy array
print("s['b']:", s['b'])            # label-based access
print("s[1]:  ", s.iloc[1])         # position-based access
```

```python
# DataFrame — the one you use 99% of the time
df = pd.DataFrame({
    'name':   ['Alice', 'Bob', 'Carol', 'Dave', 'Eve'],
    'age':    [25, 30, 27, 35, 28],
    'city':   ['Delhi', 'Mumbai', 'Delhi', 'Bangalore', 'Mumbai'],
    'salary': [60000, 80000, 70000, 95000, 72000],
    'score':  [88.5, 92.0, 78.3, 95.1, 81.7]
})

print(df)
```

## 2. Loading Data

```python
# Reading CSV — the most common format
# pd.read_csv('file.csv')

# Key parameters you should know:
# pd.read_csv('data.csv',
#     index_col=0,           # use first column as index
#     header=0,              # row to use as column names (default 0)
#     usecols=['col1','col2'], # load only specific columns
#     dtype={'age': int},    # force column type
#     parse_dates=['date'],  # parse as datetime
#     na_values=['NA', '?'], # additional strings to treat as NaN
#     nrows=1000,            # only load first 1000 rows
#     sep=';',               # separator for non-comma files
# )

# For this notebook, we create data inline
# In real projects, replace this with pd.read_csv(...)

data = {
    'PassengerId': range(1, 11),
    'Survived':    [0, 1, 1, 1, 0, 0, 0, 0, 1, 1],
    'Pclass':      [3, 1, 3, 1, 3, 3, 1, 3, 3, 2],
    'Name':        ['Braund, Mr. Owen', 'Cumings, Mrs. John', 'Heikkinen, Miss. Laina',
                    'Futrelle, Mrs. Jacques', 'Allen, Mr. William', 'Moran, Mr. James',
                    'McCarthy, Mr. Timothy', 'Palsson, Master. Gosta', 'Johnson, Mrs. Oscar',
                    'Nasser, Mrs. Nicholas'],
    'Sex':         ['male','female','female','female','male','male','male','male','female','female'],
    'Age':         [22, 38, 26, 35, 35, None, 54, 2, 27, 14],
    'Fare':        [7.25, 71.28, 7.92, 53.10, 8.05, 8.46, 51.86, 21.07, 11.13, 30.07],
}
df = pd.DataFrame(data)
print(df)
```

## 3. Inspecting a DataFrame

These are the first commands you run on any new dataset. Run them in this order.

```python
print("--- shape ---")
print(df.shape)              # (rows, cols)

print("\n--- dtypes ---")
print(df.dtypes)

print("\n--- head ---")
print(df.head(3))            # first 3 rows

print("\n--- tail ---")
print(df.tail(3))            # last 3 rows
```

```python
print("--- info ---")
df.info()                    # non-null counts + dtypes — best for spotting missing data
```

```python
print("--- describe (numeric columns) ---")
print(df.describe())

print("\n--- describe (all columns including object) ---")
print(df.describe(include='all'))
```

```python
print("--- missing values per column ---")
print(df.isnull().sum())

print("\n--- missing values percentage ---")
print((df.isnull().sum() / len(df) * 100).round(2))

print("\n--- unique values per column ---")
print(df.nunique())

print("\n--- value counts for a column ---")
print(df['Sex'].value_counts())
print(df['Sex'].value_counts(normalize=True).round(3))   # proportions
```

## 4. Selecting Data — loc vs iloc

**`loc`** — label-based. Uses actual column names and index labels. Inclusive on both ends.

**`iloc`** — position-based. Uses integer positions. End is exclusive (like Python slicing).


```python
# Select columns
print("Single column (Series):")
print(df['Age'])

print("\nMultiple columns (DataFrame):")
print(df[['Name', 'Age', 'Survived']])
```

```python
# loc — label based
print("df.loc[0:3, 'Name':'Sex']")    # rows 0 to 3 (inclusive), cols Name to Sex (inclusive)
print(df.loc[0:3, 'Name':'Sex'])
```

```python
# iloc — position based
print("df.iloc[0:3, 0:3]")           # rows 0,1,2 and cols 0,1,2
print(df.iloc[0:3, 0:3])

print("\ndf.iloc[-3:, -2:]")         # last 3 rows, last 2 columns
print(df.iloc[-3:, -2:])
```

```python
# Single value — use .at[] (label) or .iat[] (position) for speed
print("df.at[2, 'Name']:  ", df.at[2, 'Name'])
print("df.iat[2, 3]:      ", df.iat[2, 3])
```

## 5. Filtering Rows

```python
# Boolean condition
print("Survivors:")
print(df[df['Survived'] == 1][['Name', 'Sex', 'Age']])
```

```python
# Multiple conditions — use & (and), | (or), ~ (not)
# IMPORTANT: wrap each condition in parentheses
print("Female survivors:")
print(df[(df['Survived'] == 1) & (df['Sex'] == 'female')][['Name', 'Age']])

print("\nPclass 1 OR Fare > 50:")
print(df[(df['Pclass'] == 1) | (df['Fare'] > 50)][['Name', 'Pclass', 'Fare']])
```

```python
# .query() — cleaner syntax for complex conditions
print(df.query('Age > 25 and Survived == 1')[['Name', 'Age']])

# .isin() — filter by list of values
print("\nPclass 1 or 2:")
print(df[df['Pclass'].isin([1, 2])][['Name', 'Pclass']])

# .between()
print("\nAge between 20 and 30:")
print(df[df['Age'].between(20, 30)][['Name', 'Age']])
```

## 6. Adding & Modifying Columns

```python
df2 = df.copy()   # always work on a copy when modifying

# Add new column
df2['FarePerClass'] = df2['Fare'] / df2['Pclass']
print("New column:")
print(df2[['Name', 'Fare', 'Pclass', 'FarePerClass']].head())
```

```python
# Conditional column using np.where
df2['AgeGroup'] = np.where(df2['Age'] < 18, 'child', 'adult')
print(df2[['Name', 'Age', 'AgeGroup']])
```

```python
# pd.cut — bin continuous values into categories
df2['FareBin'] = pd.cut(df2['Fare'],
                         bins=[0, 15, 50, 200],
                         labels=['low', 'mid', 'high'])
print(df2[['Name', 'Fare', 'FareBin']])
```

```python
# Rename columns
df3 = df2.rename(columns={'Sex': 'gender', 'Pclass': 'passenger_class'})
print("Renamed columns:", df3.columns.tolist())

# Drop columns
df3 = df3.drop(columns=['AgeGroup', 'FareBin', 'FarePerClass'])
print("After drop:", df3.columns.tolist())
```

## 7. Handling Missing Values

```python
df4 = df.copy()
print("Missing before:\n", df4.isnull().sum())
```

```python
# dropna — use carefully, you lose data
df_dropped = df4.dropna()                    # drop any row with a null
df_dropped_col = df4.dropna(axis=1)          # drop any column with a null
df_thresh = df4.dropna(thresh=8)             # keep rows with at least 8 non-nulls
print("Rows after dropna:", len(df_dropped))
```

```python
# fillna — preferred in ML
df4['Age'] = df4['Age'].fillna(df4['Age'].median())   # fill with median (robust to outliers)
print("Missing after fillna:\n", df4.isnull().sum())
print("\nAges after fill:", df4['Age'].tolist())
```

```python
# Forward fill / backward fill — useful for time series
s = pd.Series([1, None, None, 4, None, 6])
print("Forward fill:", s.ffill().tolist())
print("Backward fill:", s.bfill().tolist())
```

## 8. Sorting & Ranking

```python
# Sort by one column
print(df.sort_values('Fare', ascending=False)[['Name', 'Fare']].head())

# Sort by multiple columns
print("\nSort by Pclass then Age:")
print(df.sort_values(['Pclass', 'Age'])[['Name', 'Pclass', 'Age']])
```

```python
# Rank
df5 = df.copy()
df5['FareRank'] = df5['Fare'].rank(ascending=False)
print(df5[['Name', 'Fare', 'FareRank']].sort_values('FareRank'))
```

## 9. Groupby & Aggregation

```python
# Basic groupby
print("Survival rate by sex:")
print(df.groupby('Sex')['Survived'].mean())

print("\nMean fare by class:")
print(df.groupby('Pclass')['Fare'].mean().round(2))
```

```python
# Multiple aggregations with .agg()
result = df.groupby('Sex').agg(
    total_passengers = ('PassengerId', 'count'),
    survival_rate    = ('Survived', 'mean'),
    avg_age          = ('Age', 'mean'),
    avg_fare         = ('Fare', 'mean')
).round(2)
print(result)
```

```python
# groupby + transform — adds result back as a column (same length as original)
df6 = df.copy()
df6['MeanFareByClass'] = df6.groupby('Pclass')['Fare'].transform('mean')
print(df6[['Name', 'Pclass', 'Fare', 'MeanFareByClass']].sort_values('Pclass'))
```

## 10. Merging & Joining

```python
# Two small DataFrames to merge
passengers = pd.DataFrame({
    'PassengerId': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Carol', 'Dave', 'Eve'],
    'Survived': [0, 1, 1, 0, 1]
})

cabin_info = pd.DataFrame({
    'PassengerId': [1, 2, 3, 6],    # note: 4,5 missing; 6 extra
    'Cabin': ['C101', 'B45', 'C101', 'D22']
})

# inner — only matching rows
print("INNER merge:")
print(pd.merge(passengers, cabin_info, on='PassengerId', how='inner'))

print("\nLEFT merge (keep all passengers):")
print(pd.merge(passengers, cabin_info, on='PassengerId', how='left'))
```

```python
# concat — stack DataFrames
df_part1 = passengers.iloc[:3]
df_part2 = passengers.iloc[3:]

combined = pd.concat([df_part1, df_part2], axis=0, ignore_index=True)
print("Concatenated:\n", combined)
```

## 11. Apply & Map

```python
df7 = df.copy()

# .map() — element-wise transformation on a Series
df7['SexEncoded'] = df7['Sex'].map({'male': 0, 'female': 1})
print(df7[['Sex', 'SexEncoded']])
```

```python
# .apply() on a Series — applies a function to each element
df7['FareLog'] = df7['Fare'].apply(lambda x: np.log1p(x))
print(df7[['Fare', 'FareLog']].round(3))
```

```python
# .apply() on a DataFrame — applies a function to each row (axis=1) or column (axis=0)
def age_category(row):
    if pd.isnull(row['Age']):
        return 'unknown'
    elif row['Age'] < 18:
        return 'child'
    else:
        return 'adult'

df7['AgeCategory'] = df7.apply(age_category, axis=1)
print(df7[['Name', 'Age', 'AgeCategory']])
```

## 12. String Operations

```python
# .str accessor — vectorized string methods
df8 = df.copy()

print(df8['Name'].str.upper().head())
print("\nContains 'Mrs':")
print(df8['Name'].str.contains('Mrs.'))
print("\nExtract title:")
df8['Title'] = df8['Name'].str.extract(r',\s([A-Za-z]+)\.')
print(df8[['Name', 'Title']])
```

## 13. Common Gotchas

```python
# GOTCHA 1: SettingWithCopyWarning
# Wrong — modifying a slice of a DataFrame
df_wrong = df[df['Survived'] == 1]
# df_wrong['NewCol'] = 1   # This raises SettingWithCopyWarning

# Correct — use .copy() or .loc
df_correct = df[df['Survived'] == 1].copy()
df_correct['NewCol'] = 1
print("Correct approach — no warning:", df_correct['NewCol'].tolist())
```

```python
# GOTCHA 2: Chained assignment with loc
df9 = df.copy()
# Wrong: df9[df9['Age'] > 25]['Survived'] = 0   # doesn't modify original!

# Correct:
df9.loc[df9['Age'] > 25, 'Survived'] = 0
print("Used .loc to modify correctly")
print(df9[df9['Age'] > 25][['Name', 'Age', 'Survived']])
```

```python
# GOTCHA 3: loc is inclusive on both ends, iloc is not
df10 = df.copy()
print("loc[0:3] — rows 0,1,2,3 (4 rows):")
print(df10.loc[0:3, ['Name']])

print("\niloc[0:3] — rows 0,1,2 (3 rows):")
print(df10.iloc[0:3][['Name']])
```

```python
# GOTCHA 4: groupby result index
result = df.groupby('Sex')['Survived'].mean()
print("Groupby result (Sex is the index):")
print(result)
print("\nType:", type(result))

# To get a DataFrame with Sex as a regular column:
result_df = df.groupby('Sex', as_index=False)['Survived'].mean()
print("\nWith as_index=False:")
print(result_df)
```
